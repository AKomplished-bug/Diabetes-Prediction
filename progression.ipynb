```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("sugar_level_progression.csv")
```

```python
# Excluding non-numeric columns before calculating mean, median, and standard deviation
numeric_data = data.drop(columns=['id', 'first_name', 'gender'])

mean_sugar_levels = numeric_data.mean(axis=0)
median_sugar_levels = numeric_data.median(axis=0)
std_sugar_levels = numeric_data.std(axis=0)

print("Mean Sugar Levels:")
print(mean_sugar_levels)
print("\nMedian Sugar Levels:")
print(median_sugar_levels)
print("\nStandard Deviation of Sugar Levels:")
print(std_sugar_levels)
```

```python
numeric_data = data.drop(columns=['id', 'first_name', 'gender'])

mean_values = numeric_data.mean(axis=0)[3:]
median_values = numeric_data.median(axis=0)[3:]

# Set up the figure and axis for plotting
plt.figure(figsize=(10, 6))

# Plot mean values
plt.plot(mean_values, label='Mean', marker='o')
# Plot median values
plt.plot(median_values, label='Median', marker='o')

# Add labels and title
plt.xlabel("Month")
plt.ylabel("Sugar Level")
plt.title("Mean and Median Sugar Levels Over Months")
plt.xticks(range(len(mean_values)), mean_values.index, rotation=45)
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
